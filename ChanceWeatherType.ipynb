```python
import pandas as pd
```

```python
df = pd.read_excel('data3.xlsx')
```

```python
df.head()
```

```python
df['athlete_gender'] = df['athlete_gender'].replace({'мужской':1, 'женский':0})
```

```python
df['rain.1h'] = df['rain.1h'].fillna(0)
df['snow.1h'] = df['snow.1h'].fillna(0)
```

```python
df.head()
```

```python
df.info()
```

```python
av_time = df.groupby(['event_name']).agg(
    {'record_id': 'count', 'finish_time': 'mean', 'athlete_gender': 'mean', 'athlete_age': 'mean', 
     'feels_like': 'mean', 'humidity': 'mean', 'clouds': 'mean', 'wind_speed': 'mean', 
     'rain.1h': 'mean', 'snow.1h': 'mean' }).reset_index()
av_time.columns = ['место забега', 'кол-во забегов', 'среднее время забегов', 'пол бегунов', 'средний возраст бегунов',
                   'средняя fells_like температура', 'средняя влажность', 'средняя облачность', 'средняя скорость ветра',
                  'средняя частота дождя', 'средняя частота снега']
av_time
```

```python
weather_by_city = df.groupby(['event_name', 'weather.main']).size().reset_index(name='counts')
weather_by_city
```

```python
weather_types = df['weather.main'].unique()

weather_counts = df.groupby(['event_name', 'weather.main']).size().reset_index(name='counts')

# Шаг 3: Подсчитываем общее количество записей для каждого города
total_counts = df.groupby('event_name').size().reset_index(name='total')

# Шаг 4: Объединяем таблицы погоды
merged_weather = pd.merge(weather_counts, total_counts, on='event_name')

# Шаг 5: Рассчитываем проценты
merged_weather['percentage'] = (merged_weather['counts'] / merged_weather['total']) * 100

# Шаг 6: Создаём полную комбинацию городов и типов погоды
full_combinations = pd.MultiIndex.from_product([df['event_name'].unique(), weather_types], names=['event_name', 'weather.main']).to_frame(index=False)

# Шаг 7: Объединяем таблицу с полной комбинацией и рассчитанные данные по погоде
weather_full = pd.merge(full_combinations, merged_weather[['event_name', 'weather.main', 'percentage']], on=['event_name', 'weather.main'], how='left')

# Заполняем отсутствующие проценты нулями
weather_full['percentage'] = weather_full['percentage'].fillna(0)

weather_full.columns = ['место забега', 'вид погоды', 'вероятность']

weather_full
```

```python
av_time_full = pd.merge(av_time, weather_full, on='место забега', how='inner') 
av_time_full
```

```python
av_time_full.to_excel('av_table.xlsx')
```
