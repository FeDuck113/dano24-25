```python
import pandas as pd
import random as rd
```

```python
df = pd.read_excel('data3.xlsx')
```

```python
df.info()
```

```python
men_df = df[df['athlete_gender'] == 'мужской'] 
men_df
```

```python
women_df = df[df['athlete_gender'] == 'женский']
women_df
```

```python
topFast_men = men_df.groupby('event_name').agg({'finish_time': 'mean'}).reset_index()
topFast_men = topFast_men.sort_values(by='finish_time')
topFast_men.columns = ['event_name', 'city_mean_time'] 
topFast_cities_men = topFast_men.head(5)['event_name'].unique()
topFast_cities_men
```

```python
topSlow_men = men_df.groupby('event_name').agg({'finish_time': 'mean'}).reset_index()
topSlow_men = topSlow_men.sort_values(by='finish_time', ascending=[0])
topSlow_men.columns = ['event_name', 'city_mean_time'] 
topSlow_cities_men = topSlow_men.head(5)['event_name'].unique()
topSlow_cities_men
```

```python
topFast_women = women_df.groupby('event_name').agg({'finish_time': 'mean'}).reset_index()
topFast_women = topFast_women.sort_values(by='finish_time')
topFast_women.columns = ['event_name', 'city_mean_time'] 
topFast_cities_women = topFast_women.head(5)['event_name'].unique()
topFast_cities_women
```

```python
topSlow_women = women_df.groupby('event_name').agg({'finish_time': 'mean'}).reset_index()
topSlow_women = topSlow_women.sort_values(by='finish_time', ascending=[0])
topSlow_women.columns = ['event_name', 'city_mean_time'] 
topSlow_cities_women = topSlow_women.head(5)['event_name'].unique()
topSlow_cities_women
```

```python
dfFast_men = df[df['event_name'].isin(topFast_cities_men)]
dfFast_men = dfFast_men.sort_values(by='event_name')
dfFastFull_men = pd.merge(dfFast_men, topFast_men, on='event_name')
dfFastFull_men
```

```python
dfSlow_men = df[df['event_name'].isin(topSlow_cities_men)]
dfSlow_men = dfSlow_men.sort_values(by='event_name')
dfSlowFull_men = pd.merge(dfSlow_men, topSlow_men, on='event_name')
dfSlowFull_men
```

```python
dfFast_women = df[df['event_name'].isin(topFast_cities_women)]
dfFast_women = dfFast_women.sort_values(by='event_name')
dfFastFull_women = pd.merge(dfFast_women, topFast_women, on='event_name')
dfFastFull_women
```

```python
dfSlow_women = df[df['event_name'].isin(topSlow_cities_women)]
dfSlow_women = dfSlow_women.sort_values(by='event_name')
dfSlowFull_women = pd.merge(dfSlow_women, topSlow_women, on='event_name')
dfSlowFull_women
```

```python
dfFastFull_men.to_excel('fast_men.xlsx')
dfSlowFull_men.to_excel('slow_men.xlsx')
dfFastFull_women.to_excel('fast_women.xlsx')
dfSlowFull_women.to_excel('slow_women.xlsx')
```
